Cleaning the Discription column of each role

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('all')

# Initialize tools
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Cleaning function
def clean_text(text):
    if pd.isnull(text):
        return ""
    # Lowercase
    text = text.lower()
    # Remove punctuation and digits
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize
    words = word_tokenize(text)
    # Remove stopwords and lemmatize
    cleaned = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return " ".join(cleaned)

# Apply to Description column
data_set["Cleaned Description"] = data_set["Description"].apply(clean_text)

# Optionally, combine fields again for embeddings
data_set["cleaned_full_text"] = data_set["Assessment Name"].fillna('') + ". " + data_set["Cleaned Description"]

# Save to new file
data_set.to_csv("../data/shl_cleaned_data.csv", index=False)
print("✅ Cleaned data saved to: shl_cleaned_data.csv")
```

Creating embaddings of cleaned disripton for each role

```python
from sentence_transformers import SentenceTransformer
import json

# Load JSON
with open("../data/shl_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Add embeddings
for item in data:
    text = item.get("cleaned_full_text", "")
    embedding = model.encode(text).tolist()
    item["embedding"] = embedding

# Save to new file
with open("../data/shl_data_with_embeddings.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4)
```

Result Generation Sorting on the basis of Cosine Similarity

```python
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import json

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Load your dataset with embeddings
with open("../data/shl_data_with_embeddings.json", "r", encoding="utf-8") as f:
    data = json.load(f)

def get_top_matches(prompt, top_n=10):
    # Embed the prompt
    prompt_embedding = model.encode(prompt).reshape(1, -1)

    # Store similarity scores
    scored_results = []
    for item in data:
        test_embedding = np.array(item["embedding"]).reshape(1, -1)
        similarity = cosine_similarity(prompt_embedding, test_embedding)[0][0]
        scored_results.append((similarity, item))

    # Sort by similarity
    top_matches = sorted(scored_results, key=lambda x: x[0], reverse=True)[:top_n]

    # Format results
    results = []
    for score, item in top_matches:
        results.append({
            "Assessment Name": item["Assessment Name"],
            "URL": item["URL"],
            "Similarity Score": round(score, 3),
            "Job Level": item.get("Job Level", ""),
            "Test Type": item.get("Test Type", ""),
            "Remote Testing": item.get("Remote Testing", ""),
            "Duration (mins)": item.get("Duration", ""),
            "Description": item["Description"]
        })

    return results
```

```python
# Example HR prompt
prompt = "Looking for a test for a marketing team lead"
results = get_top_matches(prompt)

# Display
for res in results:
    print(f"🔹 {res['Assessment Name']} (Score: {res['Similarity Score']})")
    print(f"   🔗 {res['URL']}")
    print(f"   🧠 Level: {res['Job Level']} | ⏱ Duration: {res['Duration (mins)']}")
    print(f"   📄 {res['Description']}")
    print()
```
